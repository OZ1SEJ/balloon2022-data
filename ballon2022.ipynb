```
#!/usr/bin/env python3
import pandas as pd
import matplotlib.pyplot as plt
import numpy,sys

df = pd.read_csv("ballon2022data.csv")
df.head()

df["latdiff"] = df.lat.diff()
df["londiff"] = df.lon.diff()

df.latdiff = df.latdiff * 111111.11111
df.londiff = df.londiff * 111111.11111 * numpy.cos(df.lat)

df["horiz_diff"] = ( df.latdiff**2 + df.londiff**2 ) ** 0.5

df["timediff"] = df.millis.diff()
df["altdiff"]  = df.presalt.diff()

df["horiz_vel"] = df.horiz_diff / df.timediff * 1000
df["verti_vel"] = df.altdiff    / df.timediff * 1000

plt.plot(df.millis/1000,df.pres)
plt.grid()
plt.title("Lufttryk som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("BMP180 tryk / [Pa]")
plt.show()

plt.plot(df.millis/1000,df.temp)
plt.grid()
plt.title("Temperatur som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("BMP180 temperatur / [°C]")
plt.show()

plt.plot(df.millis/1000,df.presalt,label="BMP180 trykhøjde")
plt.plot(df.millis/1000,df.gpsalt,label="GPS højde")
plt.grid()
plt.title("Højde som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Højde / [m]")
plt.show()

plt.plot(df.millis/1000,df.vbat)
plt.grid()
plt.title("Regulatorspænding som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Regulatorspænding / [V]")
plt.show()

plt.plot(df.millis/1000,df.ax/32768*2.0,label="MPU-6050 $a_x$")
plt.plot(df.millis/1000,df.ay/32768*2.0,label="MPU-6050 $a_y$")
plt.plot(df.millis/1000,df.az/32768*2.0,label="MPU-6050 $a_z$")
plt.grid()
plt.legend()
plt.title("Acceleration som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Acceleration / [G]")
plt.show()

plt.plot(df.millis/1000,df.gx/32768*2000.0,label="MPU-6050 $g_x$")
plt.plot(df.millis/1000,df.gy/32768*2000.0,label="MPU-6050 $g_y$")
plt.plot(df.millis/1000,df.gz/32768*2000.0,label="MPU-6050 $g_z$")
plt.grid()
plt.legend()
plt.title("Rotation som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Rotationshastighed / [°/s]")
plt.show()

plt.plot(df.millis/1000,df.horiz_vel)
plt.grid()
plt.title("Horisontal hastighed som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Horisontal hastighed / [m/s]")
plt.show()

plt.plot(df.horiz_vel,df.presalt)
plt.grid()
plt.title("Horisontal hastighed som funktion af trykhøjde")
plt.xlabel("Horisontal hastighed / [m/s]")
plt.ylabel("Trykhøjde / [m]")
plt.ylim([0,12000])
plt.show()

plt.plot(df.millis/1000,df.verti_vel)
plt.grid()
plt.title("Vertikal hastighed som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Vertikal hastighed / [m/s]")
plt.show()

plt.plot(df.verti_vel,df.presalt)
plt.grid()
plt.title("Vertikal hastighed som funktion af trykhøjde")
plt.xlabel("Vertikal hastighed / [m/s]")
plt.ylabel("Trykhøjde / [m]")
plt.show()
```
