```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ballon2022data.csv")
df.head()

plt.plot(df.millis/1000,df.pres)
plt.grid()
plt.title("Lufttryk som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("BMP180 tryk / [Pa]")
plt.show()

plt.plot(df.millis/1000,df.temp)
plt.grid()
plt.title("Temperatur som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("BMP180 temperatur / [°C]")
plt.show()

plt.plot(df.millis/1000,df.presalt,label="BMP180 trykhøjde")
plt.plot(df.millis/1000,df.gpsalt,label="GPS højde")
plt.grid()
plt.title("Højde som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Højde / [m]")
plt.show()

plt.plot(df.millis/1000,df.vbat)
plt.grid()
plt.title("Regulatorspænding som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Regulatorspænding / [V]")
plt.show()

plt.plot(df.millis/1000,df.ax/32768*2.0,label="MPU-6050 $a_x$")
plt.plot(df.millis/1000,df.ay/32768*2.0,label="MPU-6050 $a_y$")
plt.plot(df.millis/1000,df.az/32768*2.0,label="MPU-6050 $a_z$")
plt.grid()
plt.legend()
plt.title("Acceleration som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Acceleration / [G]")
plt.show()

plt.plot(df.millis/1000,df.gx/32768*2000.0,label="MPU-6050 $g_x$")
plt.plot(df.millis/1000,df.gy/32768*2000.0,label="MPU-6050 $g_y$")
plt.plot(df.millis/1000,df.gz/32768*2000.0,label="MPU-6050 $g_z$")
plt.grid()
plt.legend()
plt.title("Rotation som funktion af tid")
plt.xlabel("Tid / [s]")
plt.ylabel("Rotationshastighed / [°/s]")
plt.show()

```
